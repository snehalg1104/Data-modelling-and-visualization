```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv("AirQuality.csv", encoding='iso-8859-1')

df

relavant_variables=df.columns

relavant_variables

```

```python
# Create a line plot for overall AQI trends

plt.figure(figsize=(12, 6))

sns.lineplot(data=df['AQI Value'])

plt.title('Overall AQI Trend')

plt.xlabel('Data Point')

plt.ylabel('AQI Value')

plt.grid(True)

plt.show()
```

```python
# Create a line plot for PM2.5 AQI trends

plt.figure(figsize=(12, 6))

sns.lineplot(data=df['PM2.5 AQI Value'])

plt.title('PM2.5 AQI Trend')

plt.xlabel('Data Point')

plt.ylabel('PM2.5 AQI Value')

plt.grid(True)

plt.show()
```
